```python
from datetime import datetime, timedelta
from zoneinfo import ZoneInfo

import json

def encontrar_intervalo(fecha, intervalos):
    return next((i for i in intervalos if i['Fecha_inicio'] <= fecha <= i['Fecha_fin']), None)

with open('datos_modulos.json', 'r', encoding='utf-8') as f:
    datos_modulos = json.load(f)

trucks_input = datos_modulos["body"]["trucks_input"]
avances = datos_modulos["body"]["avances"]
delta = datos_modulos["body"]["delta"]

for avance in avances:
    avance['Fecha'] = datetime.strptime(avance['Fecha'], '%Y-%m-%d %H:%M:%S').replace(tzinfo=ZoneInfo('America/Santiago'))

# Generar intervalos en una lÃ­nea
intervalos = [
    {
        'Intervalo': f"{a['Avance']} - {b['Avance']}", 
        'Fecha_inicio': a['Fecha'], 
        'Fecha_fin': b['Fecha']
    } 
    for a, b in zip(avances, avances[1:])
]
```

```python
for truck_record in trucks_input:
    time_empty_UTC = datetime.strptime(truck_record['time_empty'], '%Y-%m-%dT%H:%M:%SZ').replace(tzinfo=ZoneInfo('UTC'))
    time_empty_stacking = time_empty_UTC + timedelta(hours=delta)
    intervalo = encontrar_intervalo(time_empty_stacking, intervalos)

    intervalo_string = intervalo['Intervalo']
    
    truck_record['intervalo'] = intervalo_string
```
